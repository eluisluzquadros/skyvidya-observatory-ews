```python
# RAG Geoespacial com Kepler.gl no Google Colab
# Execute cada célula em sequência

# ==============================================================================
# CÉLULA 1: Instalação de Dependências
# ==============================================================================

# !pip install geopandas pandas numpy shapely
# !pip install langchain openai
# !pip install keplergl
# !pip install duckdb duckdb-engine
# !pip install sentence-transformers
# !pip install chromadb
# !pip install ipywidgets
# !jupyter nbextension enable --py --sys-prefix keplergl
# !pip install -U langchain-community

# ==============================================================================
# CÉLULA 2: Imports e Configurações
# ==============================================================================

import os
import json
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import Point
import warnings
warnings.filterwarnings('ignore')

from google.colab import output
output.disable_custom_widget_manager()

# Kepler.gl
from keplergl import KeplerGl

# LangChain
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings, HuggingFaceEmbeddings
from langchain.llms import OpenAI
from langchain.schema import Document
from langchain.text_splitter import CharacterTextSplitter

# DuckDB para queries espaciais
import duckdb

# Configurar API Key da OpenAI (opcional - pode usar modelo local)
os.environ["OPENAI_API_KEY"] = "SUA_OPENAI_API_KEY_AQUI"

print("✅ Dependências instaladas e importadas com sucesso!")

# ==============================================================================
# CÉLULA 3: Criação de Dados de Exemplo
# ==============================================================================

def create_sample_geodata():
    """
    Criar dataset de exemplo com cidades brasileiras e dados de risco climático
    """
    cidades_data = {
        'nome': [
            'São Paulo', 'Rio de Janeiro', 'Brasília', 'Salvador',
            'Fortaleza', 'Belo Horizonte', 'Manaus', 'Curitiba',
            'Recife', 'Goiânia', 'Belém', 'Porto Alegre',
            'Guarulhos', 'Campinas', 'São Luís', 'São Gonçalo',
            'Maceió', 'Duque de Caxias', 'Natal', 'Teresina',
            'Campo Grande', 'Nova Iguaçu', 'São Bernardo do Campo', 'João Pessoa',
            'Osasco', 'Santo André', 'Jaboatão dos Guararapes', 'Contagem'
        ],
        'lat': [
            -23.5505, -22.9068, -15.7942, -12.9714, -3.7172,
            -19.9167, -3.1190, -25.4244, -8.0476, -16.6799,
            -1.4558, -30.0346, -23.4628, -22.9056, -2.5307,
            -22.8305, -9.6658, -22.7858, -5.8130, -5.0892,
            -20.4486, -22.7592, -23.6914, -7.1195, -23.5329,
            -23.6586, -8.1137, -19.9167
        ],
        'lon': [
            -46.6333, -43.1729, -47.8825, -38.5014, -38.5434,
            -43.9345, -60.0217, -49.2654, -34.8770, -49.2639,
            -48.4682, -51.2177, -46.5336, -47.0608, -44.3068,
            -43.2096, -35.7353, -43.3114, -35.2044, -42.8019,
            -54.6295, -43.4509, -46.5647, -34.8641, -46.7916,
            -46.5386, -35.0134, -44.0537
        ],
        'populacao': [
            12325232, 6747815, 3055149, 2886698, 2669342,
            2521564, 2219580, 1963726, 1653461, 1516113,
            1499641, 1488252, 1392121, 1213792, 1108975,
            1084839, 1025360, 924624, 890480, 868075,
            906092, 804815, 833240, 817511, 696050,
            721368, 702621, 668949
        ],
        'tendencia_risco': [
            'estável', 'crescente', 'decrescente', 'crescente',
            'crescente', 'estável', 'crescente', 'decrescente',
            'crescente', 'estável', 'crescente', 'decrescente',
            'estável', 'decrescente', 'crescente', 'crescente',
            'crescente', 'estável', 'crescente', 'crescente',
            'decrescente', 'estável', 'estável', 'crescente',
            'estável', 'decrescente', 'crescente', 'estável'
        ],
        'qualidade_vida': [
            7.2, 6.8, 8.1, 6.5, 6.3, 7.5, 5.9, 8.3,
            6.1, 7.0, 5.8, 7.8, 7.1, 7.9, 6.2, 6.4,
            6.7, 6.9, 6.8, 6.0, 7.4, 6.6, 7.3, 6.5,
            7.2, 7.6, 6.3, 7.1
        ],
        'risco_climatico': [
            6.5, 7.8, 4.2, 8.1, 8.9, 5.5, 7.3, 3.8,
            8.5, 5.2, 7.9, 4.1, 6.7, 4.5, 8.2, 7.6,
            8.0, 7.1, 7.9, 8.3, 4.8, 7.2, 6.8, 7.7,
            6.9, 4.9, 8.1, 5.8
        ],
        'regiao': [
            'Sudeste', 'Sudeste', 'Centro-Oeste', 'Nordeste', 'Nordeste',
            'Sudeste', 'Norte', 'Sul', 'Nordeste', 'Centro-Oeste',
            'Norte', 'Sul', 'Sudeste', 'Sudeste', 'Nordeste',
            'Sudeste', 'Nordeste', 'Sudeste', 'Nordeste', 'Nordeste',
            'Centro-Oeste', 'Sudeste', 'Sudeste', 'Nordeste', 'Sudeste',
            'Sudeste', 'Nordeste', 'Sudeste'
        ]
    }

    df = pd.DataFrame(cidades_data)

    # Calcular score de segurança baseado na tendência e qualidade de vida
    df['score_seguranca'] = df.apply(lambda row:
        (3 if row['tendencia_risco'] == 'decrescente' else
         2 if row['tendencia_risco'] == 'estável' else 1) * row['qualidade_vida']
    , axis=1)

    # Criar geometria
    geometry = [Point(xy) for xy in zip(df.lon, df.lat)]
    gdf = gpd.GeoDataFrame(df, geometry=geometry, crs='EPSG:4326')

    return gdf

# Criar dados
gdf_cidades = create_sample_geodata()
print(f"✅ Dataset criado com {len(gdf_cidades)} cidades")
print("\n📊 Primeiras 5 cidades:")
print(gdf_cidades[['nome', 'tendencia_risco', 'qualidade_vida', 'score_seguranca']].head())

# ==============================================================================
# CÉLULA 4: Classe RAG Geoespacial
# ==============================================================================

class GeoRAGEngine:
    def __init__(self, gdf, use_openai=False):
        self.gdf = gdf.copy()
        self.use_openai = use_openai

        # Configurar embeddings (local ou OpenAI)
        if use_openai and "OPENAI_API_KEY" in os.environ:

            self.embeddings = OpenAIEmbeddings()
            self.llm = OpenAI(temperature=0)
        else:
            # Usar modelo local (HuggingFace)
            self.embeddings = HuggingFaceEmbeddings(
                model_name='sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'
            )
            self.llm = None  # Usaremos lógica baseada em regras

        self.setup_vector_store()
        self.setup_spatial_db()
        print("✅ RAG Engine inicializado")

    def setup_vector_store(self):
        """Criar embeddings das descrições das cidades"""
        texts = []
        metadatas = []

        for idx, row in self.gdf.iterrows():
            text = f"""
            Cidade: {row['nome']}
            Região: {row['regiao']}
            População: {row['populacao']:,} habitantes
            Tendência de risco climático: {row['tendencia_risco']}
            Qualidade de vida: {row['qualidade_vida']}/10
            Risco climático atual: {row['risco_climatico']}/10
            Score de segurança: {row['score_seguranca']:.2f}
            """
            texts.append(text.strip())
            metadatas.append({"index": idx, "nome": row['nome']})

        # Criar vector store
        documents = [Document(page_content=text, metadata=meta)
                    for text, meta in zip(texts, metadatas)]

        self.vectorstore = Chroma.from_documents(
            documents=documents,
            embedding=self.embeddings,
            persist_directory=None
        )
        print(f"✅ Vector store criado com {len(texts)} documentos")

    def setup_spatial_db(self):
        """Configurar DuckDB para queries espaciais"""
        self.conn = duckdb.connect(':memory:')

        # MODIFICAÇÃO: Converter a coluna 'geometry' para WKT (string)
        # para que o DuckDB possa entendê-la.
        df_for_db = self.gdf.copy()
        df_for_db['geometry'] = df_for_db['geometry'].astype(str)

        # Registrar o DataFrame modificado no DuckDB
        self.conn.register('cidades', df_for_db)
        print("✅ Database espacial configurado")

    def parse_query_rules(self, query):
        """Parser simples baseado em regras (fallback sem LLM)"""
        query_lower = query.lower()

        # Detectar tipo de consulta
        if any(word in query_lower for word in ['top', 'melhores', 'primeiras']):
            tipo = 'ranking'
        elif any(word in query_lower for word in ['região', 'estado', 'próximo']):
            tipo = 'filtro_espacial'
        else:
            tipo = 'busca_geral'

        # Detectar critério
        if any(word in query_lower for word in ['segura', 'segurança', 'seguro']):
            criterio = 'seguranca'
        elif any(word in query_lower for word in ['qualidade', 'vida', 'viver']):
            criterio = 'qualidade_vida'
        elif any(word in query_lower for word in ['risco', 'clima', 'climático']):
            criterio = 'risco_climatico'
        else:
            criterio = 'geral'

        # Detectar número
        import re
        numbers = re.findall(r'\d+', query)
        limite = int(numbers[0]) if numbers else 10

        return {
            'tipo': tipo,
            'criterio': criterio,
            'limite': limite,
            'query_original': query
        }

    def execute_spatial_query(self, parsed_query):
        """Executar query espacial no DuckDB"""
        # (código original com uma pequena modificação para reconstruir o GeoDataFrame)
        criterio = parsed_query['criterio']
        limite = parsed_query['limite']

        if criterio == 'seguranca':
            # Ordenar por score de segurança (decrescente + qualidade de vida)
            sql = """
            SELECT * FROM cidades
            ORDER BY score_seguranca DESC, qualidade_vida DESC
            LIMIT ?
            """
        elif criterio == 'qualidade_vida':
            sql = """
            SELECT * FROM cidades
            ORDER BY qualidade_vida DESC, score_seguranca DESC
            LIMIT ?
            """
        elif criterio == 'risco_climatico':
            sql = """
            SELECT * FROM cidades
            ORDER BY risco_climatico ASC, qualidade_vida DESC
            LIMIT ?
            """
        else:
            # Query geral - balanceada
            sql = """
            SELECT * FROM cidades
            ORDER BY (score_seguranca + qualidade_vida - risco_climatico) DESC
            LIMIT ?
            """

        result_df = self.conn.execute(sql, [limite]).fetchdf()

        # MODIFICAÇÃO: Reconstruir o GeoDataFrame a partir dos resultados
        # do DuckDB, garantindo que a geometria seja preservada.
        # Usamos o índice original do GeoDataFrame para selecionar as linhas corretas.
        result_gdf = self.gdf.loc[result_df.index].copy()

        return result_gdf.head(limite)

    def generate_response_rules(self, query, spatial_results):
        """Gerar resposta usando regras (sem LLM)"""
        n_results = len(spatial_results)

        if 'segura' in query.lower():
            intro = f"🛡️ **Top {n_results} cidades mais seguras para morar:**\n\n"
            criterio_texto = "segurança (tendência de risco decrescente + qualidade de vida)"
        elif 'qualidade' in query.lower():
            intro = f"🏆 **Top {n_results} cidades com melhor qualidade de vida:**\n\n"
            criterio_texto = "qualidade de vida"
        else:
            intro = f"📊 **Top {n_results} cidades recomendadas:**\n\n"
            criterio_texto = "score combinado (segurança + qualidade de vida)"

        response_text = intro

        for idx, (_, cidade) in enumerate(spatial_results.iterrows(), 1):
            response_text += f"""
**{idx}. {cidade['nome']} ({cidade['regiao']})**
- 🏘️ População: {cidade['populacao']:,} habitantes
- 📈 Tendência de risco: {cidade['tendencia_risco']}
- ⭐ Qualidade de vida: {cidade['qualidade_vida']}/10
- 🌡️ Risco climático atual: {cidade['risco_climatico']}/10
- 🛡️ Score de segurança: {cidade['score_seguranca']:.2f}

"""

        response_text += f"\n💡 **Critério de ordenação:** {criterio_texto}"

        # Adicionar insights
        regioes = spatial_results['regiao'].value_counts()
        if len(regioes) > 1:
            response_text += f"\n\n🗺️ **Distribuição por região:** {dict(regioes)}"

        return response_text

    def prepare_kepler_data(self, spatial_results, query_info):
        """Preparar dados para visualização no Kepler.gl"""
        # Converter para formato compatível com Kepler
        df_map = spatial_results.copy()
        df_map['longitude'] = df_map.geometry.x
        df_map['latitude'] = df_map.geometry.y
        df_map['rank'] = range(1, len(df_map) + 1)

        # Adicionar categoria de cor baseada na tendência
        color_map = {
            'decrescente': [34, 197, 94],    # Verde - mais seguro
            'estável': [245, 158, 11],       # Amarelo - moderado
            'crescente': [239, 68, 68]       # Vermelho - mais risco
        }

        df_map['color_rgb'] = df_map['tendencia_risco'].map(color_map)

        # Configuração do Kepler.gl
        config = {
            "version": "v1",
            "config": {
                "mapState": {
                    "latitude": -15.7942,
                    "longitude": -47.8825,
                    "zoom": 4
                },
                "mapStyle": {
                    "styleType": "dark"
                },
                "visState": {
                    "layers": [
                        {
                            "id": "cidades",
                            "type": "point",
                            "config": {
                                "dataId": "cidades",
                                "label": "Cidades Seguras",
                                "color": [255, 255, 255],
                                "columns": {
                                    "lat": "latitude",
                                    "lng": "longitude"
                                },
                                "isVisible": True,
                                "visConfig": {
                                    "radius": 15,
                                    "fixedRadius": False,
                                    "opacity": 0.8,
                                    "outline": True,
                                    "thickness": 2,
                                    "radiusRange": [10, 50],
                                    "colorRange": {
                                        "name": "Custom",
                                        "type": "custom",
                                        "category": "Custom",
                                        "colors": ["#22c55e", "#f59e0b", "#ef4444"]
                                    }
                                }
                            }
                        }
                    ],
                    "interactionConfig": {
                        "tooltip": {
                            "fieldsToShow": {
                                "cidades": [
                                    "nome", "regiao", "populacao",
                                    "tendencia_risco", "qualidade_vida",
                                    "score_seguranca", "rank"
                                ]
                            }
                        }
                    }
                }
            }
        }

        return df_map[['nome', 'regiao', 'populacao', 'tendencia_risco',
                      'qualidade_vida', 'risco_climatico', 'score_seguranca',
                      'longitude', 'latitude', 'rank']], config

    def query(self, user_question):
        """Pipeline completo do RAG geoespacial"""
        print(f"🔍 Processando: '{user_question}'")

        # 1. Parse da query
        parsed_query = self.parse_query_rules(user_question)
        print(f"📋 Detectado: {parsed_query}")

        # 2. Executar busca espacial
        spatial_results = self.execute_spatial_query(parsed_query)
        print(f"📍 Encontradas {len(spatial_results)} cidades")

        # 3. Gerar resposta textual
        text_response = self.generate_response_rules(user_question, spatial_results)

        # 4. Preparar dados para Kepler
        kepler_data, kepler_config = self.prepare_kepler_data(spatial_results, parsed_query)

        return {
            'text_response': text_response,
            'spatial_results': spatial_results,
            'kepler_data': kepler_data,
            'kepler_config': kepler_config,
            'total_results': len(spatial_results)
        }

# Inicializar RAG Engine
rag_engine = GeoRAGEngine(gdf_cidades, use_openai=False)

# ==============================================================================
# CÉLULA 5: Função de Interface do Usuário
# ==============================================================================

def consultar_rag_geoespacial(pergunta):
    """
    Função principal para fazer consultas ao RAG geoespacial
    """
    print("=" * 80)
    print(f"🗺️ RAG GEOESPACIAL - CONSULTA")
    print("=" * 80)

    # Processar query
    resultado = rag_engine.query(pergunta)

    # Mostrar resposta textual
    print("\n📝 RESPOSTA:")
    print("-" * 40)
    print(resultado['text_response'])

    # Criar mapa Kepler.gl
    print(f"\n🗺️ CRIANDO MAPA COM {resultado['total_results']} RESULTADOS...")

    mapa = KeplerGl(height=600, data={'cidades': resultado['kepler_data']})

    # Aplicar configuração personalizada se disponível
    if resultado['kepler_config']:
        try:
            mapa.config = resultado['kepler_config']
        except:
            pass  # Usar configuração padrão se houver erro

    return mapa, resultado

# ==============================================================================
# CÉLULA 6: Exemplos de Uso
# ==============================================================================

# Exemplo 1: Cidades mais seguras
print("🔥 EXEMPLO 1: Cidades mais seguras")
mapa1, resultado1 = consultar_rag_geoespacial("Quais as top 10 cidades mais seguras para morar?")

# Mostrar o mapa
mapa1

# ==============================================================================
# CÉLULA 7: Mais Exemplos de Consultas
# ==============================================================================

# Exemplo 2: Qualidade de vida
print("\n" + "="*50)
print("🔥 EXEMPLO 2: Melhor qualidade de vida")
mapa2, resultado2 = consultar_rag_geoespacial("Top 8 cidades com melhor qualidade de vida")
mapa2

# ==============================================================================
# CÉLULA 8: Consulta Interativa
# ==============================================================================

# Interface interativa para novas consultas
def nova_consulta():
    """Interface para fazer novas consultas"""
    print("\n" + "="*60)
    print("🤖 FAÇA SUA PRÓPRIA CONSULTA")
    print("="*60)
    print("Exemplos de perguntas:")
    print("• 'Top 5 cidades mais seguras'")
    print("• 'Cidades com menor risco climático'")
    print("• 'Melhores cidades para viver'")
    print("• 'Top 15 cidades recomendadas'")
    print("-"*60)

    pergunta = input("Digite sua pergunta: ")

    if pergunta.strip():
        mapa, resultado = consultar_rag_geoespacial(pergunta)
        return mapa
    else:
        print("❌ Pergunta vazia!")
        return None

# Descomentar a linha abaixo para usar a interface interativa
# mapa_custom = nova_consulta()

# ==============================================================================
# CÉLULA 9: Visualização de Estatísticas
# ==============================================================================

import matplotlib.pyplot as plt

def mostrar_estatisticas(resultado):
    """Mostrar estatísticas dos resultados"""
    df = resultado['spatial_results']

    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('📊 Estatísticas das Cidades Selecionadas', fontsize=16, fontweight='bold')

    # Gráfico 1: Distribuição por região
    df['regiao'].value_counts().plot(kind='bar', ax=axes[0,0], color='skyblue')
    axes[0,0].set_title('Distribuição por Região')
    axes[0,0].set_xlabel('Região')
    axes[0,0].set_ylabel('Número de Cidades')
    axes[0,0].tick_params(axis='x', rotation=45)

    # Gráfico 2: Tendência de risco
    df['tendencia_risco'].value_counts().plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%')
    axes[0,1].set_title('Tendência de Risco Climático')

    # Gráfico 3: Qualidade de vida vs Risco climático
    axes[1,0].scatter(df['qualidade_vida'], df['risco_climatico'],
                     c=df['score_seguranca'], cmap='RdYlGn', s=100)
    axes[1,0].set_xlabel('Qualidade de Vida')
    axes[1,0].set_ylabel('Risco Climático')
    axes[1,0].set_title('Qualidade de Vida vs Risco Climático')

    # Gráfico 4: Score de segurança
    # MODIFICAÇÃO: Usar head(10) em vez de nsmallest, já que 'rank' não está no DataFrame aqui.
    df_top_10 = df.head(10)
    df_top_10[['nome', 'score_seguranca']].plot(
        x='nome', y='score_seguranca', kind='bar', ax=axes[1,1], color='green', alpha=0.7)
    axes[1,1].set_title('Score de Segurança (Top Cidades)')
    axes[1,1].set_xlabel('Cidade')
    axes[1,1].set_ylabel('Score')
    axes[1,1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

    # Tabela resumo
    print("\n📋 TABELA RESUMO:")
    print("-" * 50)
    colunas_resumo = ['nome', 'regiao', 'qualidade_vida', 'score_seguranca', 'tendencia_risco']
    print(df[colunas_resumo].head(10).to_string(index=False))

# Mostrar estatísticas do último resultado
mostrar_estatisticas(resultado1)

# ==============================================================================
# CÉLULA 10: Exportar Dados
# ==============================================================================

def exportar_resultados(resultado, nome_arquivo="resultado_rag_geoespacial"):
    """Exportar resultados para diferentes formatos"""
    df = resultado['spatial_results']

    # CSV
    df.to_csv(f"{nome_arquivo}.csv", index=False)
    print(f"✅ Dados exportados para {nome_arquivo}.csv")

    # GeoJSON para uso em outros mapas
    df.to_file(f"{nome_arquivo}.geojson", driver="GeoJSON")
    print(f"✅ Dados geoespaciais exportados para {nome_arquivo}.geojson")

    # Relatório em texto
    with open(f"{nome_arquivo}_relatorio.txt", "w", encoding="utf-8") as f:
        f.write(resultado['text_response'])
    print(f"✅ Relatório exportado para {nome_arquivo}_relatorio.txt")

# Exportar resultados (opcional)
exportar_resultados(resultado1, "cidades_mais_seguras")

print("\n🎉 SISTEMA RAG GEOESPACIAL PRONTO!")
print("Execute as células acima em sequência para usar o sistema.")
print("Modifique os dados na CÉLULA 3 para usar seus próprios datasets.")
```
